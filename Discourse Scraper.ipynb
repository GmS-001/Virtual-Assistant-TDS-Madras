```python
from selenium import webdriver
import time
import requests
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
wait = WebDriverWait(driver, 4)
actions = ActionChains(driver)
url = 'https://tds.s-anand.net/#/2025-01/'
driver.get(url)
wait.until(EC.new_window_is_opened)
```

```python
# Collection of all urls from main site
sidebar_list = driver.find_element(By.XPATH,'/html/body/main/aside/div[2]')
chapter_list = sidebar_list.find_elements(By.XPATH, './ul/li')
print((sidebar_list))
url_list = []
for item in chapter_list :
    if item.get_attribute('class') == 'folder level-1' :
        print('Getting all the topics of Chatper: ',item.text)
        a_tag = item.find_element(By.XPATH,'./a')
        url_list.append(a_tag.get_attribute('href'))
        topics = item.find_elements(By.XPATH, './/ul/li')
        for topic in topics :
            a_inside = topic.find_element(By.XPATH,'./a')
            url_list.append(a_inside.get_attribute('href'))
    elif item.get_attribute('class') == 'file' :
        print('Chapter: ',item.text)
        a_tag = item.find_element(By.XPATH,'./a')
        url_list.append(a_tag.get_attribute('href'))
print('Extracted all url list with length : ',len(url_list))
driver.quit()
print('Driver is closed.')
```

```python
def extract_visible_text_from_site(url, scroll_pause=2, max_scrolls=10):
    # Scroll to load dynamic content
    
    last_height = driver.execute_script("return document.body.scrollHeight")
    for _ in range(max_scrolls):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(scroll_pause)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    # Extract all visible text from all tags (excluding scripts/styles)
    elements = driver.find_elements(By.XPATH, "//*[not(self::script or self::style)]")

    text_parts = []
    for el in elements:
        text = el.text.strip()
        if text:
            text_parts.append(text)

    

    # Combine and clean text
    full_text = "\n".join(text_parts)
    return full_text
```

```python
data = ''
for url in url_list :
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get(url)
    print(f'Page {driver.title} opened.')
    text = extract_visible_text_from_site(url)
    data = data + text
    driver.quit()
    print('Driver is closed.')
print('Data extracted successfully.')
```

```python
# Collect data from all sites
def extract_data(url_list) :
    data = ''
    for url in url_list :
        driver = intialize_driver(url)
        text = extract_visible_text_from_site(driver, url)
        data = data + text
    print('Data extracted successfully.')
    return data
```

```python
def extract_urls(driver) :
    # Collection of all urls from main site
    sidebar_list = driver.find_element(By.XPATH,'/html/body/main/aside/div[2]')
    chapter_list = sidebar_list.find_elements(By.XPATH, './ul/li')
    print((sidebar_list))
    url_list = []
    for item in chapter_list :
        if item.get_attribute('class') == 'folder level-1' :
            print('Getting all the topics of Chatper: ',item.text)
            a_tag = item.find_element(By.XPATH,'./a')
            url_list.append(a_tag.get_attribute('href'))
            topics = item.find_elements(By.XPATH, './/ul/li')
            for topic in topics :
                a_inside = topic.find_element(By.XPATH,'./a')
                url_list.append(a_inside.get_attribute('href'))
        elif item.get_attribute('class') == 'file' :
            print('Chapter: ',item.text)
            a_tag = item.find_element(By.XPATH,'./a')
            url_list.append(a_tag.get_attribute('href'))
    print('Extracted all url list with length : ',len(url_list))
    driver.quit()
    print('Driver is closed.')
    return url_list
```

```python
def intialize_driver(url) :
    # Setup headless Chrome
    options = Options()
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")

    driver = webdriver.Chrome(options=options)
    driver.get(url)
    print(f'Driver : {driver.title} is opened.')
    return driver
```

```python
url = 'https://tds.s-anand.net/#/2025-01/'
driver = intialize_driver(url)
url_list = extract_urls(driver)
document = extract_data(url_list)
```
